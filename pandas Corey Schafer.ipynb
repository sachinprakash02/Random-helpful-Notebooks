#### Video 9 : Data Cleaning

```python
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
df3 = pd.DataFrame(people)
df3
```

```python
df3.dropna() # removes 3,4,5 rows
```

```python
df3.dropna(axis='index',how="any") # same as above ; how = "all" ,
```

```python
df3.dropna(axis="columns",how="all")
```

```python
df3.dropna(axis="index",how="any",subset=["email"]) # dropping from specific column
```

```python
df3.replace("NA",np.nan,inplace=True) # replacing a paticular occurance
```

```python
df3.replace("Missing",np.nan,inplace=True)
```

```python
df3
```

```python
df3.isna()
```

```python
df3.fillna("MISSING") # filling np.nan with "MISSING"
```

```python
# type casting :
df3.dtypes
```

```python
type(np.nan)
```

```python
df3["age"].fillna(0,inplace=True) # or directly type cast to float since np.nan is float
```

```python
df3["age"] =df3["age"].astype('int')
```

```python
df3.dtypes
```

```python
df3["age"] =df3["age"].astype('float')
df3.dtypes
```

```python
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
```

```python
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
df = pd.read_csv('ETH_1hr.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df.loc[0,"Date"]
```

```python
df.loc[0,"Date"].day_name()
```

```python
df["Date"].dt.day_name().head()
```

```python
df["DayOfWeek"] = df["Date"].dt.day_name()
```

```python
df.head()
```

```python
filt = (df["Date"] >= pd.to_datetime('2019-01-01')) & (df["Date"] <= '2020')
df.loc[filt].shape
```

```python
df.set_index("Date",inplace=True)
```
